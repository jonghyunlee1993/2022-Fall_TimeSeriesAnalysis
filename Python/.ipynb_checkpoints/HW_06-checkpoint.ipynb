```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
from statsmodels.tsa.arima_process import arma_generate_sample

rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False

ar_coefs = [1, 0.5]
ma_coefs = [1]

Z = arma_generate_sample(ar_coefs, ma_coefs, nsample=100)
plt.title("그림 6-1 AR procss phi=0.5")
plt.plot(Z)
plt.hlines(0, 0, 100, "k")
plt.show()
```

```python
ar_coefs = [1, -0.5]
ma_coefs = [1]

Z = arma_generate_sample(ar_coefs, ma_coefs, nsample=100)
plt.title("그림 6-2 AR procss phi=-0.5")
plt.plot(Z)
plt.hlines(0, 0, 100, "k")
plt.show()
```
